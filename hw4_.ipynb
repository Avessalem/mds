```python
import numpy as np
```

### Уровень 0

```python
def func(x):
    return x**5+4*np.sin(2*x)+np.cos(3*x+3)
```

```python
from scipy.misc import derivative as drv
```

```python
print('Производная функции в точке х=1 ', drv(func, 1, n=1))
print('Вторая производная функции в точке х=1 ', drv(func, 1, n=2))
```

### Уровень 1

```python
import matplotlib.pyplot as plt
```

```python
def func_1(x):
    return (np.sin(2*x+1))**5
```

```python
x = np.linspace(-5,5)
fx = func_1(x)
flx = [derivative(func_1, xx, n=1) for xx in x]
```

```python
plt.plot(x,fx,label = 'function')
plt.plot(x,flx, label = 'derivative')
plt.legend()
plt.grid()
plt.show()
```
