### Уровень 0

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix = np.array([range(1,10)]).reshape(3,3) #Создаём матрицу
matrix
```

```python
matrix = matrix + np.eye(3, 3) #Прибавляем единичную матрицу
print(matrix)
```

```python
print(np.dot(matrix, np.array([1, 2, 3]))) #Умножение на вектор
```

```python
print(np.linalg.inv(matrix)) #Обратная матрица
```

### Уровень 1

```python
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
'Уменьшаем стороны фигуры в 2 раза'

x = (A/2)[:,0] 
y = (A/2)[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
'Поворачиваем фигуру на 130 градусов против часовой стрелки'

rot_matrix = np.array([[np.cos(np.deg2rad(130)), - np.sin(np.deg2rad(130))], [np.sin(np.deg2rad(130)), np.cos(np.deg2rad(130))]])
A_rotated = A[:, 0:2] @ rot_matrix
x = A_rotated[:,0] 
y = A_rotated[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
'Отражаем фигуру относительно y=x'

x = A[:,0] 
y = A[:,1]

x1 = A[:,1] 
y1 = A[:,0]

plt.plot(x, y, lw = 3, c = 'y')
plt.plot(x1, y1, linestyle ='--', c='r')
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.legend(['A','A_mirrored'])
plt.grid()
```

```python
'Отражаем фигуру относительно y=x при помощи умножения на матрицу'

x = A @ np.array([[0, 1, 0],[1, 0, 0],[0, 0, 1]])[:,0]
y = A @ np.array([[0, 1, 0],[1, 0, 0],[0, 0, 1]])[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```

```python
'Смещаем фигуру на (200,300)'

A_moved = A[:,0:2] + np.array([200,300])


x = A_moved[:,0] 
y = A_moved[:,1]

plt.plot(x, y)
plt.ylim([0, 500])
plt.xlim([0, 500])
plt.grid()
```

```python
'Растягиваем фигуру на (0.5, 1.5)'

A_stretched = A[:,0:2] @ np.array([[0.5, 0],[0, 1.5]])

x = A_stretched[:,0] 
y = A_stretched[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.grid()
```
